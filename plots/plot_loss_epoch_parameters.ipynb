```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# no da
inputs, dp = {}, {}
rms = pd.read_csv('../outputs/parameters/logger_01-06-2021_age_[0-70]_RMS_wd_0_L_dp0.0_no_da_clr_[-4.,-3.57].csv').iloc[:-1].astype('float')
inputL = rms[rms['epoch']<200]
inputL['dat'] = 'Without DtAugmentation: training'
inputL['dav'] = 'Without DtAugmentation: validation'
# inputL['da'] = 'Without Data Augmentation'
# da
rms = pd.read_csv('../outputs/parameters/logger_01-06-2021_age_[0-70]_RMS_wd_0_L_dp0.0_clr_[-5,-2.9].csv').iloc[:-1].astype('float')
inputR = rms[rms['epoch']<200]
inputR['dat'] = 'With DtAugmentation: training'
inputR['dav'] = 'With DtAugmentation: validation'
# inputR['da'] = 'Data Augmentation'
inputs = pd.concat([inputL, inputR])

dt = inputs
sns.set(context='paper', style='whitegrid',
        font_scale=1.8, color_codes=True,
        rc={'figure.figsize':(14,7)})
for pt in [True, False]:
    f, ax = plt.subplots()

    if pt:
        sns.lineplot(x='epoch', y='loss', data=dt, hue='dat',
                     palette='Set2')
        sns.lineplot(x='epoch', y='val_loss', data=dt, hue='dav',
                     palette='Set1')
        ax.set(xlabel='Épocas', ylabel='Função de perda (MAE)')
        handles, labels = plt.gca().get_legend_handles_labels()
        labels = ['Sem DtAugmentation: treinamento',
                  'Com DtAugmentation: treinamento',
                  'Sem DtAugmentation: validação',
                  'Com DtAugmentation: validação']
        order = [1, 3, 0, 2]
        plt.legend([handles[idx] for idx in order],[labels[idx] for idx in order],
                    loc='upper right',
                    fancybox=True)
    else:
        sns.lineplot(x='epoch', y='loss', data=dt, hue='dat',
                     palette='Set2')
        sns.lineplot(x='epoch', y='val_loss', data=dt, hue='dav',
                     palette='Set1')
        ax.set(xlabel='Epoch', ylabel='Loss (MAE)')
        handles, labels = plt.gca().get_legend_handles_labels()
        order = [1, 3, 0, 2]
        plt.legend([handles[idx] for idx in order],[labels[idx] for idx in order],
                    loc='upper right',
                    fancybox=True)
        

    ax.set_ylim(3, 18)
    ax.set_xlim(60, 150)
    ax.axvline(202, 0, 600, color='red', linestyle='--')
    plt.tight_layout()
    name = 'Loss_da_and_no_da'
    print(name)
    if pt:
        f.savefig(name+".png")
    else:
        f.savefig(name+"_us.png")
```

```python
def smooth(data, weight=0.6):
    scalar = data
    last = scalar[0]
    smoothed = []
    for point in scalar:
        smoothed_val = last * weight + (1 - weight) * point
        smoothed.append(smoothed_val)
        last = smoothed_val
    return smoothed

# no da
inputs, dp = {}, {}
rms = pd.read_csv('../outputs/parameters/logger_22-06-2021_age_[0-70]_RMS_wd_0_L_dp0.0_no_da_clr_[-4.3,-4]2.csv').iloc[:-1].astype('float')
inputL = rms[rms['epoch']<200]

# inputL['loss'] = smooth(inputL['loss'])
# inputL['val_loss'] = smooth(inputL['val_loss'])
inputL['dat'] = 'Without DtAugmentation: training'
inputL['dav'] = 'Without DtAugmentation: validation'
# inputL['da'] = 'Without Data Augmentation'
# da
rms = pd.read_csv(
    '../outputs/parameters/logger_21-06-2021_age_[0-70]_RMS_wd_0_L_dp0.0_clr_[-5.2,-3.4].csv').iloc[:-1].astype('float')
inputR = rms[rms['epoch']<200]
# inputL['loss'] = smooth(inputL['loss'])
# inputL['val_loss'] = smooth(inputL['val_loss'])
inputR['dat'] = 'With DtAugmentation: training'
inputR['dav'] = 'With DtAugmentation: validation'
# inputR['da'] = 'Data Augmentation'
inputs = pd.concat([inputL, inputR])

dt = inputs
sns.set(context='paper', style='whitegrid',
        font_scale=1.8, color_codes=True,
        rc={'figure.figsize':(14,7)})
for pt in [True, False]:
    f, ax = plt.subplots()

    if pt:
        sns.lineplot(x='epoch', y='loss', data=dt, hue='dat',
                     palette='Set2')
        sns.lineplot(x='epoch', y='val_loss', data=dt, hue='dav',
                     palette='Set1')
        ax.set(xlabel='Épocas', ylabel='Função de perda (MAE)')
        handles, labels = plt.gca().get_legend_handles_labels()
        labels = ['Sem AumentoDados: treinamento',
                  'Com AumentoDados: treinamento',
                  'Sem AumentoDados: validação',
                  'Com AumentoDados: validação']
        order = [1, 3, 0, 2]
        plt.legend([handles[idx] for idx in order],[labels[idx] for idx in order],
                    loc='upper right',
                    fancybox=True)
    else:
        sns.lineplot(x='epoch', y='loss', data=dt, hue='dat',
                     palette='Set2')
        sns.lineplot(x='epoch', y='val_loss', data=dt, hue='dav',
                     palette='Set1')
        ax.set(xlabel='Epoch', ylabel='Loss (MAE)')
        handles, labels = plt.gca().get_legend_handles_labels()
        order = [1, 3, 0, 2]
        plt.legend([handles[idx] for idx in order],[labels[idx] for idx in order],
                    loc='upper right',
                    fancybox=True)
        

    ax.set_ylim(0, 50)
    ax.set_xlim(6, 150)
    ax.axvline(202, 0, 600, color='red', linestyle='--')
    plt.tight_layout()
    name = 'Loss_da_and_no_da'
    print(name)

    if pt:
        f.savefig(name+".png")
    else:
        f.savefig(name+"_us.png")
```
