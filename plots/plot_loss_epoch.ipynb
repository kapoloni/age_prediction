```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set(context='paper', style='whitegrid',
        font_scale=1.8, color_codes=True,
        rc={'figure.figsize':(16,8)})
```

```python
# left
inputs, dp = {}, {}
rms = pd.read_csv('../outputs/logger_07-05-2021_age_[0-70]_RMS_wd_0_L_dp0.4.csv').iloc[:-1].astype('float')
inputs['RMSL'] = rms[rms['epoch']<600]
dp['L'] = '0.6'
# right
rms = pd.read_csv('../outputs/logger_07-05-2021_age_[0-70]_RMS_wd_0_L_dp0.2.csv').iloc[:-1].astype('float')
inputs['RMSR'] = rms[rms['epoch']<600]
dp['R'] = '0.3'

# for opt in ['Adam', 'SGD', 'SGDm', 'RMS']:
for opt in ['RMS']:
    for side in ['L', 'R']:
        dt = inputs[opt+side]
#         for pt in [True, False]:
        for pt in [False]:

            f, ax = plt.subplots()

            if pt:
                sns.lineplot(x='epoch', y='loss', data=dt,
                             label='Treinamento', color=sns.color_palette('Set1')[0])
                sns.lineplot(x='epoch', y='val_loss', data=dt,
                             label='Validao', color=sns.color_palette('Set1')[1])
                ax.set(xlabel='pocas', ylabel='Funo de perda (MSE)')
            else:
                sns.lineplot(x='epoch', y='loss', data=dt,
                             label='Training', color=sns.color_palette('Set1')[0])
                sns.lineplot(x='epoch', y='val_loss', data=dt,
                             label='Validation', color=sns.color_palette('Set1')[1])
                ax.set(xlabel='Epoch', ylabel='Loss (MSE)')

            ax.set_ylim(0, 20)
            ax.set_xlim(6, 600)
            ax.axvline(202, 0, 600, color='red', linestyle='--')
            # plt.xticks([10**x for x in range(-10, 1, 2)])
            plt.tight_layout()
            name = 'Loss_'+side+'_'+opt+'_0_70_dp'+dp[side]
            print(name)
            if pt:
                f.savefig(name+".png")
            else:
                f.savefig(name+"_us.png")
        # break
```

```python
# left
inputs = {}
adam = pd.read_csv('outputs/logger_19-03-2021_age_[0-70]_Adam_wd_0_L.csv')
inputs['AdamL'] = adam[adam['epoch']<300]
sgd = pd.read_csv('outputs/logger_20-03-2021_age_[0-70]_SGD_wd_0_L.csv')
inputs['SGDL'] = sgd[sgd['epoch']<300]
sgdm = pd.read_csv('outputs/logger_24-03-2021_age_[0-70]_SGDm_wd_0_L.csv')
inputs['SGDmL'] = sgdm[sgdm['epoch']<300]
rms = pd.read_csv('outputs/logger_20-03-2021_age_[0-70]_RMS_wd_0_L.csv')
inputs['RMSL'] = rms[rms['epoch']<300]
# right
adam = pd.read_csv('outputs/logger_21-03-2021_age_[0-70]_Adam_wd_0_R.csv')
inputs['AdamR'] = adam[adam['epoch']<300]
sgd = pd.read_csv('outputs/logger_21-03-2021_age_[0-70]_SGD_wd_0_R.csv')
inputs['SGDR'] = sgd[sgd['epoch']<300]
sgdm = pd.read_csv('outputs/logger_25-03-2021_age_[0-70]_SGDm_wd_0_R.csv')
inputs['SGDmR'] = sgdm[sgdm['epoch']<300]
rms = pd.read_csv('outputs/logger_22-03-2021_age_[0-70]_RMS_wd_0_R.csv')
inputs['RMSR'] = rms[rms['epoch']<300]
```

```python
for opt in ['Adam', 'SGD', 'SGDm', 'RMS']:
    for side in ['L', 'R']:
        dt = inputs[opt+side]
        for pt in [True, False]:

            f, ax = plt.subplots()

            if pt:
                sns.lineplot(x='epoch', y='loss', data=dt, label='Treinamento')
                sns.lineplot(x='epoch', y='val_loss', data=dt, label='Validao')
                ax.set(xlabel='pocas', ylabel='Funo de perda')
            else:
                sns.lineplot(x='epoch', y='loss', data=dt, label='Training')
                sns.lineplot(x='epoch', y='val_loss', data=dt, label='Validation')
                ax.set(xlabel='Epoch', ylabel='Loss')

            ax.set_ylim(0, 30)
            ax.set_xlim(0, 302)
            # plt.xticks([10**x for x in range(-10, 1, 2)])
            plt.tight_layout()
            name = 'Loss_'+side+'_'+opt+'_0_70'
            print(name)
            if pt:
                f.savefig("outputs_exp/imgs/"+name+".png")
            else:
                f.savefig("outputs_exp/imgs/"+name+"_us.png")
```
